```python
# 1. 

import matplotlib.pyplot as plt
from matplotlib import rc
import csv

x = []
y = []   
z = 201712

f = open("cafe_2year.csv",'r', encoding='utf-8')
reader = csv.reader(f) 
header = next(reader)

for row in reader:
    if z % 3 == 0 :
        x.append(row[0])
        y.append(row[1])
    z += 1

rc('font', family='Malgun Gothic')
font1 = { 'color':'green','size':25 }

plt.title('전국의 월별 카페수', fontdict=font1, weight="bold")
plt.xlabel('년월')
plt.ylabel('카페수')

plt.bar(x, y)
plt.show()
```

```python
# 5.

import matplotlib.pyplot as plt
from matplotlib import rc
import csv

f5 = open('doctor2.csv', 'r', encoding='utf-8')
reader2 = csv.reader(f5)

header = next(reader2)

x = []
y1 = []
y2 = []
y3 = []
y4 = []

for line in reader2:
    x.append(line[0])
    y1.append(int(line[1]))
    y2.append(int(line[2]))
    y3.append(int(line[3]))
    y4.append(int(line[4]))

rc('font', family='Malgun Gothic')
font1 = {'size':22, 'color':'green'}

plt.title('서울 및 광역시 종합병원 의사 수', fontdict=font1, weight="bold")
plt.xlabel('시')
plt.ylabel('의사 수')

plt.plot(x, y1, label='일반의', color='r', linestyle='-', marker='o')
plt.plot(x, y2, label='인턴', color='b', linestyle='--', marker='x')
plt.plot(x, y3, label='레지던트', color='yellow', linestyle=':', marker='s')
plt.plot(x, y4, label='전문의', color='purple', linestyle='-.', marker='d')
plt.legend(loc='best')

plt.show()
        
f5.close()
```

```python
# 6. 

f6 = open('doctor2.csv', 'r', encoding='utf-8')
reader3 = csv.reader(f6)
header = next(reader3)

people = ['일반의', '인턴', '레지던트', '전문의']
number = []
a = next(reader3)
b = 0
for i in range(4) :
    b += int(a[i+1])
    
for i in range(4) :
    persent = int(a[i+1])/b;
    number.append(persent)
    
font1 = {'size':22, 'color':'green'}
plt.pie(number, explode=(0, 0.1, 0, 0), labels=people, autopct='%.1f%%', shadow=True, startangle=90)
plt.title('서울 종합병원 의사 유형별 분포', fontdict=font1, weight="bold")
plt.show()
        
f6.close()
```
