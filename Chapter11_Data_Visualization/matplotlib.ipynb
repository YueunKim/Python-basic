```python
# 12-5. 

import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

xdata = ['김조은', '김하은', '김유은']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1, label='국어', color='orange', linestyle='-', marker='o')
plt.plot(xdata, ydata2, label='영어', color='#000fff', linestyle='--', marker='x')
plt.plot(xdata, ydata3, label='수학', color='red', linestyle='-.', marker='s')
plt.plot(xdata, ydata4, label='과학', color='#444444', linestyle=':', marker='d')

plt.title('세 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

- color: blue, green, red, cyan, magenta, yellow, black(k), white
- marker: circle(o), 삼각형(^), 역삼각형(v), 사각형(s), 별(*), 플러스(+)
- linestyle: 직선(-), dashed(--), dash-dot(-.), dotted(:)


```python
plt.plot(xdata, ydata1, 'ro-', label='국어') # color = red, marker = 'o', linestyle ='-'
plt.plot(xdata, ydata2, 'cx--', label='영어', linewidth=3)
plt.plot(xdata, ydata3, 'ms-.',label='수학',linewidth=3.5, markersize=8)
plt.plot(xdata, ydata4, 'kd:',label='과학',linewidth=2, markersize=10)

plt.title('세 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
# 12-6. 

import matplotlib.pyplot as plt

xdata = list(range(30))
ydata = []

for x in xdata:
    y = 2 * x
    ydata.append(y)

plt.plot(xdata, ydata, label='y=2x')
plt.xlim(0, 35)
plt.ylim(0, 70)
plt.xticks(list(range(0, 35, 2)))
plt.yticks(list(range(0, 70, 5)))

plt.grid(True)
plt.show()
```

```python
plt.title('y=2x')
plt.plot(xdata, ydata, label='y=2x', color='magenta', linewidth=5)
plt.xlim(-1, 30)
plt.ylim(-2, 60)
plt.xticks(list(range(0, 30, 2)))
plt.yticks(list(range(0, 60, 5)))

plt.grid(True)
plt.show()
```

```python
# sin, cos function

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

x = np.arange(0,2*np.pi,0.1) 
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,x,z,linewidth=5)
plt.xlabel('x values from 0 to 2pi')  
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 2pi')
plt.legend(['sin(x)', 'cos(x)'])      
plt.grid(True)
plt.show()
```

```python
import numpy as np
xs = np.arange(0, 2*np.pi+0.01, 0.01)
ysin = np.sin(xs)
ycos = np.cos(xs)

plt.plot(xs, ysin, label = 'Sine curve')
plt.plot(xs, ycos, label = 'Cosine curve')
plt.title('삼각함수 그래프')
plt.legend()
plt.xticks(np.arange(0, 2*np.pi+0.01, np.pi/2))

plt.axhline(y=0, color = 'k', linewidth = 0.5)
plt.axhline(y=1, color = 'k', linewidth = 0.5)
plt.axhline(y=-1, color = 'k', linewidth = 0.5)
plt.axvline(x=0, color = 'k', linewidth = 0.5)
plt.axvline(x=np.pi, color = 'k', linewidth = 0.5)
plt.axvline(x=2*np.pi, color = 'k', linewidth = 0.5)
plt.show()
```

```python
# sigmoid function

xs = np.linspace(-3, 3, 100)
ys = 1 / (1+np.exp(-xs))
          
plt.plot(xs, ys)
plt.ylim(-0.1, 1.1)
plt.title('Sigmoid graph')

plt.axhline(y=0, color = 'k', linewidth = 0.5)
plt.axhline(y=1, color = 'k', linewidth = 0.5)
plt.axvline(x=0, color = 'k', linewidth = 0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmo(x):   
    a = []  
    for itr in x:   # parameter로 전달 받은 x 변수(array 만큼) 반복 실행
        a.append(1/(1+np.exp(-itr)))  # 반복 횟수는 x array만큼이며, 변경되는 값은 itr
    return a  

x = np.linspace(-3, 3, 100)  # linspace (start, end, num)
sig = sigmo(x)  # 앞에서 define했던 sigmo함수 호출 (파라메트로 위에서 생성한 x array 전달)

plt.plot(x, sig, color = 'r', linewidth = 3)  
plt.show()   
```
